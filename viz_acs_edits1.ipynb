```python
chicago_df = pd.read_csv("data/chicago_internet.csv",index_col=0,
                      parse_dates=[0])
```

```python
chicago_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="darkgrid")

Population_Size = chicago_df["total_pop"]

x=100 - chicago_df["hh_no_internet(%)"]
y=chicago_df["total_black(%)"]

g= sns.relplot(x=x, y=y, size= Population_Size,
            sizes=(0, 300), alpha=.6, palette="deep", height=6, data= chicago_df, color = 'red', aspect=1.5)

a, b = np.polyfit(x, y, 1)
plt.plot(x, a*x+b, color = 'blue')

plt.ylim(-5, None)

plt.title("Internet Access by Race - Black (2015-19)" , size =22, weight=900)
plt.xlabel('Households with Internet (%)', size=16, weight=900)
plt.ylabel('Total Black Population (%)', size=16, weight=900)



```
